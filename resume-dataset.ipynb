```python
import pandas as pd
import numpy as np
import re
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from collections import Counter
from tqdm import tqdm
import matplotlib.pyplot as plt
import pickle

# Set random seed for reproducibility
torch.manual_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Step 1: Data Cleaning
def clean_text(text):
    # Remove extra whitespace and newlines
    text = re.sub(r'\s+', ' ', text.strip())
    # Remove special characters, keep alphanumeric and basic punctuation
    text = re.sub(r'[^\w\s.,-]', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

# Load and clean the dataset
df = pd.read_csv('/kaggle/input/resume-dataset/Resume/Resume.csv')
df['Resume_str'] = df['Resume_str'].apply(clean_text)

# Encode the labels
label_encoder = LabelEncoder()
df['Category_encoded'] = label_encoder.fit_transform(df['Category'])
num_classes = len(label_encoder.classes_)
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)

# Compute class weights for imbalanced dataset
class_counts = Counter(df['Category_encoded'])
total_samples = len(df)
class_weights = torch.tensor([total_samples / (num_classes * class_counts[i]) for i in range(num_classes)], dtype=torch.float).to(device)

# Step 2: Text Vectorization with TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1, 2))
X = vectorizer.fit_transform(df['Resume_str']).toarray()
y = df['Category_encoded'].values
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

# Step 3: Split data into train, validation, and test sets
# First, split into train+val (85%) and test (15%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)
# Then, split train+val into train (70% of total) and val (15% of total)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, random_state=42, stratify=y_temp  # 0.1765 ≈ 15/(85) to get 15% of total
)

# Verify split sizes
print(f"Train size: {len(X_train)} ({len(X_train)/len(X):.2%})")
print(f"Validation size: {len(X_val)} ({len(X_val)/len(X):.2%})")
print(f"Test size: {len(X_test)} ({len(X_test)/len(X):.2%})")

# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train).to(device)
y_train_tensor = torch.LongTensor(y_train).to(device)
X_val_tensor = torch.FloatTensor(X_val).to(device)
y_val_tensor = torch.LongTensor(y_val).to(device)
X_test_tensor = torch.FloatTensor(X_test).to(device)
y_test_tensor = torch.LongTensor(y_test).to(device)

# Step 4: Create PyTorch Dataset
class ResumeDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'features': self.features[idx],
            'labels': self.labels[idx]
        }

# Create datasets
train_dataset = ResumeDataset(X_train_tensor, y_train_tensor)
val_dataset = ResumeDataset(X_val_tensor, y_val_tensor)
test_dataset = ResumeDataset(X_test_tensor, y_test_tensor)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
test_loader = DataLoader(test_dataset, batch_size=32)

# Step 5: Define the Neural Network
class TextClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super(TextClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, num_classes)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc3(x)
        return x

# Initialize model, optimizer, and loss function
input_dim = X_train.shape[1]  # Number of TF-IDF features (5000)
hidden_dim = 256
model = TextClassifier(input_dim, hidden_dim, num_classes).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss(weight=class_weights)

# Step 6: Training Loop with Early Stopping
def train_epoch(model, data_loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch in tqdm(data_loader, desc="Training"):
        features = batch['features'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(features)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, preds = torch.max(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / len(data_loader), correct / total

def evaluate(model, data_loader, criterion, device, desc="Evaluating"):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(data_loader, desc=desc):
            features = batch['features'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(features)
            loss = criterion(outputs, labels)

            total_loss += loss.item()
            _, preds = torch.max(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return total_loss / len(data_loader), correct / total, all_preds, all_labels

# Training with early stopping
num_epochs = 50  # Max epochs, early stopping will likely stop earlier
patience = 5  # Number of epochs to wait for improvement
min_delta = 0.000003  # Minimum improvement in validation loss
best_val_loss = float('inf')
patience_counter = 0
best_model_state = None

train_losses = []
train_accs = []
val_losses = []
val_accs = []

for epoch in range(num_epochs):
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc, _, _ = evaluate(model, val_loader, criterion, device, desc="Validating")

    train_losses.append(train_loss)
    train_accs.append(train_acc)
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    print(f'Epoch {epoch+1}/{num_epochs}')
    print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}')
    print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}')

    # Early stopping
    if val_loss < best_val_loss - min_delta:
        best_val_loss = val_loss
        best_model_state = model.state_dict()
        patience_counter = 0
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print(f'Early stopping triggered after {epoch+1} epochs')
        break

# Step 7: Plot Training and Validation Metrics
# Plot Loss
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(train_losses) + 1), train_losses, label='Train Loss')
plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.savefig('loss_plot.png')

# Plot Accuracy
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(train_accs) + 1), train_accs, label='Train Accuracy')
plt.plot(range(1, len(val_accs) + 1), val_accs, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)
plt.savefig('accuracy_plot.png')

# Step 8: Evaluate Best Model on Test Set
model.load_state_dict(best_model_state)
test_loss, test_acc, test_preds, test_labels = evaluate(model, test_loader, criterion, device, desc="Testing")
print(f'\nTest Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}')
print("\nTest Classification Report:")
print(classification_report(test_labels, test_preds, target_names=label_encoder.classes_))

# Save the best model
torch.save(best_model_state, 'resume_classifier_tfidf_best.pth')
```

```python
import torch
import pickle
import re
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data Cleaning Function (must match training)
def clean_text(text):
    text = re.sub(r'\s+', ' ', text.strip())
    text = re.sub(r'[^\w\s.,-]', '', text)
    text = text.lower()
    return text

# Load Label Encoder (assuming it was saved during training)
with open('label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

# Load TF-IDF Vectorizer
with open('tfidf_vectorizer.pkl', 'rb') as f:
    vectorizer = pickle.load(f)

# Load Model
class TextClassifier(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super(TextClassifier, self).__init__()
        self.fc1 = torch.nn.Linear(input_dim, hidden_dim)
        self.relu = torch.nn.ReLU()
        self.dropout = torch.nn.Dropout(0.3)
        self.fc2 = torch.nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = torch.nn.Linear(hidden_dim // 2, num_classes)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc3(x)
        return x

# Initialize model with same architecture
input_dim = 5000  # From training
hidden_dim = 256
num_classes = len(label_encoder.classes_)
model = TextClassifier(input_dim, hidden_dim, num_classes).to(device)
model.load_state_dict(torch.load('resume_classifier_tfidf_best.pth', map_location=device))
model.eval()

# Inference Function
def predict_category(resume_text):
    # Clean and vectorize the text
    cleaned_text = clean_text(resume_text)
    tfidf_vector = vectorizer.transform([cleaned_text]).toarray()
    tfidf_tensor = torch.FloatTensor(tfidf_vector).to(device)

    # Predict
    with torch.no_grad():
        output = model(tfidf_tensor)
        _, pred = torch.max(output, dim=1)
        category_id = pred.item()
        category_name = label_encoder.inverse_transform([category_id])[0]
    return category_name

# Example Usage
sample_resume = """
HR ADMINISTRATOR/MARKETING ASSOCIATE with over 15 years of experience in Hospitality and Customer Service Management. 
Focused on customer satisfaction, team management, and marketing strategies. 
Certified in Customer Loyalty and Marketing by Segment.
"""

predicted_category = predict_category(sample_resume)
print(f"Predicted Category: {predicted_category}")
```
